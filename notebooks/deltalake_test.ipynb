```python
from deltalake.writer import write_deltalake
from deltalake import DeltaTable
import pandas as pd
from pathlib import Path

# load your existing gold data
df = pd.read_parquet(sorted(Path('../storage/gold/neo').glob('neo_*.parquet'))[-1])
print(f"Loaded {len(df)} rows")
import pyarrow as pa

# convert to PyArrow table first
arrow_table = pa.Table.from_pandas(df)

# write as Delta table
write_deltalake("../storage/delta/neo", arrow_table, mode="overwrite")
print("Written as Delta table!")

# read it back
dt = DeltaTable("../storage/delta/neo")
print(f"Delta table version: {dt.version()}")
print(f"Files: {dt.files()}")
```
